```python
# Verificare se tutte le colonne contengono solo valori Null, 0 o NaN
if (null_count == len(df)).all():
    print("Il DataFrame contiene solo valori Null, 0 o NaN")
```

```python
# controllare se ci sono valori nulli
if df[['var4','var5','var3']].isnull().values.any():
    print("La colonna contiene valori nulli.")
else:
    # controllare se ci sono valori diversi da 0 e NaN
    unique_values = df['var4'].unique()
    if len(unique_values)==1 and unique_values[0] in [0, NaN]:
        print("La colonna contiene solo valori con 0 o NaN o nulli.")
    else:
        print("La colonna contiene valori diversi da 0 e NaN.")
```
